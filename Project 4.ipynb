```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("IBM HR ANALYTICS.csv")

# Step 2: Calculate Attrition Rate
total_employees = len(df)
left_employees = len(df[df['Attrition'] == 'Yes'])
attrition_rate = left_employees / total_employees
print("Total Employees:", total_employees)
print("Employees who left:", left_employees)
print("Attrition Rate: {:.2f}%".format(attrition_rate * 100))

# Step 3: Visualize Attrition by Department
dept_attrition = df[df['Attrition'] == 'Yes']['Department'].value_counts()
plt.figure(figsize=(8,5))
dept_attrition.plot(kind='bar', color='skyblue')
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees Left")
plt.xticks(rotation=45)
plt.show()

# Step 4: Average Age of Employees Leaving
avg_age_left = df[df['Attrition'] == 'Yes']['Age'].mean()
print("Average Age of Employees Leaving: {:.2f}".format(avg_age_left))

# Optional: Age Distribution of Employees Leaving
plt.figure(figsize=(8,5))
df[df['Attrition'] == 'Yes']['Age'].hist(bins=10, color='orange')
plt.title("Age Distribution of Employees Leaving")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.show()
```
